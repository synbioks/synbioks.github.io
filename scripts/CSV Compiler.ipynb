```python
import csv
import json
```

```python
#needs to run once for each cleaned dataset to be integrated into the RDF/XML files.

#with open(r"C:\Users\jjett2\Desktop\SynthBio\NER Data Products\Chemical Dataset\Chemical - Cleaned Dataset.csv", "r", encoding="utf-8") as infile:
#with open(r"C:\Users\jjett2\Desktop\SynthBio\NER Data Products\Species Dataset\Species - Cleaned and Grounded.csv", "r", encoding="utf-8") as infile:
with open(r"C:\Users\jjett2\Desktop\SynthBio\NER Data Products\Species Dataset\Strains from Species Dataset - Cleaned and Grounded.csv", "r", encoding="utf-8") as infile:
    #specific to local file directory system. Will need to refactor for local situation
    csvin = csv.reader(infile)
    headers = next(csvin)
    data = [r for r in csvin]
    
```

```python
#data
```

```python
chemIDs = {}
for row in data:
    chebi=row[0]
    pubs=row[3]
    #print(chebi, pubs)
    splitpubs=pubs.split(",")
    #print(splitpubs)
    cleanpubs=[p.strip() for p in splitpubs]
    #print(cleanpubs)
    for p in cleanpubs:
        #print(chebi,p)
        if p in chemIDs:
            chemIDs[p].append(chebi)
        else:
            chemIDs[p]=[chebi]
        #chemIDs[p]=chebi
```

```python
#chemIDs
with open(r"C:\Users\jjett2\Desktop\SynthBio\NER Data Products\Species Dataset\ncbiStrainValues.json", "w", encoding="utf-8") as outfile:
    #specific to local file directory system. Will need to refactor for local situation
    json.dump(chemIDs,outfile)
```
